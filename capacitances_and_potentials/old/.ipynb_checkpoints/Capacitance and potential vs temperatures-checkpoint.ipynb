```python
def smooth(x,window_len=11,window='hanning'):
    """http://scipy-cookbook.readthedocs.io/items/SignalSmooth.html"""
    s=np.r_[x[window_len-1:0:-1],x,x[-1:-window_len:-1]]
    #print(len(s))
    if window == 'flat': #moving average
        w=np.ones(window_len,'d')
    else:
        w=eval('np.'+window+'(window_len)')
    y=np.convolve(w/w.sum(),s,mode='valid')
    return y
```

```python
# smoothing
ws = 251

fig, ax=plt.subplots(figsize=(8,6))

plt.plot(data300[:,1], data300[:,2],label='0', alpha=.6, color=color_wheel[0])

plt.plot(data400[:,1],data400[:,2],'-', label='400',color=color_wheel[-5])
plt.plot(data450[:,1],data450[:,2],'-', label='450',color=color_wheel[-2])
plt.plot(data450npt[:,1],data450npt[:,2],'-', label='450-npt',color=color_wheel[-4])


smoothing_factor = [ 40, 250]
for i,sh in enumerate(smoothing_factor):
    # smoothing
    window = sp.signal.gaussian(ws, sh)
    # convolve the arrays
    sC_h = sp.signal.convolve(data350[:,2], window/window.sum(), mode='same')
    plt.plot(data350[:,1], sC_h, '--', label=str(sh), alpha=.6, color=color_wheel[i+1])



plt.xlim(-8,8)
#plt.ylim(-20,20)
plt.xlabel('V')
plt.ylabel('$C$')


box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
